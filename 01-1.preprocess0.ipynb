```python
import pandas as pd

df = pd.read_excel('movie_data.xlsx',sheetname = 1)
df.head()
```

```python
# 모델링에 방해되는 일부 columns 삭제
df1 = df.drop(['Name','Code','Released','SeoulAud','NoScreen'], axis = 1)
df1.head()
```

```python
# 더미변수로 바꿔야하는 columns
df2 = pd.concat([df1.ix[:,0:2],df1.ix[:,3]],axis = 1) 
df2.head()
```

```python
# 더미변수로 변환
Genre_d = pd.get_dummies(df2['Genre'])
Level_d = pd.get_dummies(df2['Level'])
Country_d = pd.get_dummies(df2['Country'])
print Genre_d.shape
print Level_d.shape
print Country_d.shape

# 더미변수로 바꾸면 변수가 3개에서 80개로 바뀐다.
```

```python
# 더미변수끼리 모으기
dummies0 = Genre_d.ix[:,:'war'].join(Level_d.ix[:,'all':])
dummies = dummies0.join(Country_d.ix[:,:'usa'])
dummies.head()
```

```python
# 원래 데이터프레임에서 카테고리값을 가진 변수를 삭제하고 더미변수 이어붙이기
df3 = df1.drop(['Genre','Level','Country'],axis =1)
df3 = df3.join(dummies)
df3
```
